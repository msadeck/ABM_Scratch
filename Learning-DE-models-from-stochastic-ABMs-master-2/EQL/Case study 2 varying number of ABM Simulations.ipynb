```python
from PDEFind_class_online import PDE_Findclass
import time, glob, os, pdb
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import set_matplotlib_formats
set_matplotlib_formats('png', 'pdf')
from model_selection_IP3 import ODE_sim,learned_RHS_ODE
from PDE_FIND3 import print_pde

#increase font size throughout
font = {'size'   : 20}
plt.rc('font', **font)
```

```python
#PDE-FIND values
#Where data is stored
data_dir = "../data/"
#Additional name in filenames
comp_str = ""
model_name = ""

#percentage of training, testing points
trainPerc = 0.5
valPerc = 0.5

#degree of library
deg = 4

#Learning algorithm
algo_name = 'Greedy'

#pruning percentage
prune_level = 0

num_eqns = 3
#save_learned_eqns = True
#save_xi = True
#save_learned_xi = True

#number of equations to learn
reals = 10
```

## Perform EQL for several rp values

```python
#Realizations of ABM Simulation we will consider
real_vec = [1,5,10,25]

rp = 0.01
rd = rp/2.0

fig = plt.figure(figsize=(9,6.5))
count = 1

## Initialize summary lists list of learned xi vectors
xi_learned_reals = []
xi_learned_means = []

#list of mean squared error (MSE) between Mean-field model and ABM data
MF_MSE = []
#list of mean squared error (MSE) between learned model and ABM data
learned_MSE = []

#counter
count = 1


for real in real_vec:
    
    xi_learned_tmp = np.zeros((10,4))
    
    for data_num in np.arange(10):

        dataset = ["logistic_ABM_sim_rp_"+str(rp)+"_rd_"+str(rd)+"_real"+str(real)
                   +"_lattice_30_dataset"+str(data_num+1)]

        #Run pipeline
        pf = PDE_Findclass(dataset,
                       comp_str,
                       model_name,
                       data_dir=data_dir,
                       reals=reals,
                       trainPerc = trainPerc,
                       valPerc = valPerc,
                       deg=deg,
                       prune_level = prune_level,
                       num_eqns=num_eqns,
                       algo_name = algo_name)

        pf.train_val_ODEFind()
        pf.list_common_eqns()

        #save learned xi value        
        xi_learned_tmp[data_num,:] = pf.inferred_xi[0]
        
        
    #save learned xi's
    xi_learned_reals.append(xi_learned_tmp)
    xi_learned_means.append(np.mean(xi_learned_tmp,axis=0))
    
    #simulate learned DE Model at mean learned parameter values
    y_inferred = ODE_sim(xi_learned_means[-1],learned_RHS_ODE,pf.t,np.array([0.05]),description=pf.description,deg=deg)
    
    if real == real_vec[0]:
        #simulate mean-field model
        y_MF = ODE_sim(np.array([rp-rd,-rp]),learned_RHS_ODE,pf.t,np.array([0.05]),description=['C^1','C^2'],deg=2)


    
    learned_MSE_tmp = 0
    MF_MSE_tmp = 0
    
    ax = fig.add_subplot(2,2,count)
    
    #Load ABM data
    for data_num in np.arange(10):

        dataset = ["logistic_ABM_sim_rp_"+str(rp)+"_rd_"+str(rd)+"_real"+str(real)
                   +"_lattice_30_dataset"+str(data_num+1)]


        mat = np.load("../data/"+dataset[0]+".npy",allow_pickle=True).item()
        ABM = mat["variables"][1]

    
        ax.plot(pf.t[::4]*(rp-rd),ABM[::4],'b.-',label = "ABM Data",markersize=2,linewidth=0.5)
    
        learned_MSE_tmp += np.linalg.norm(ABM.T-y_inferred)
        MF_MSE_tmp += np.linalg.norm(ABM.T-y_MF)

    #plot mean-field and learned model
    ax.plot(pf.t*(rp-rd),y_MF.T,'k',linewidth=2,label = "MF Model")
    ax.plot(pf.t*(rp-rd),y_inferred.T,'r--',linewidth=2,label="Learned Model")
        
    #format axes
    ax.set_title("$N$ = " +str(real))#, $P_d$ = " +str(rd)
    if count > 2:
        ax.set_xlabel("Time ($T$)")
    else:
        ax.set_xticks([])
    if count % 2 == 1:
        ax.set_ylabel("Density")

    #Save MSEs, learned xi vector
    learned_MSE.append(learned_MSE_tmp/len(pf.y)/10.0)
    MF_MSE.append(MF_MSE_tmp/len(pf.y)/10.0)

    count+=1

#plt.savefig("case_study_2_BMD_converging.pdf",format="pdf")
```

## Table of inferred Equations

```python
##Creates table that can be copied and pasted into a LaTeX table environment.

print('\\begin{tabular}{|c|l|l|c|}')
print('    \\hline')
print(r'    $N$  & Mean-Field Model (MSE) & Learned Model (MSE) & $\hat{\xi}$ MSE \\ ')
print('    \\hline')

for i,real in enumerate(real_vec):

    MF = print_pde(np.array([rp-rd,-rp])[:,np.newaxis],["C","C^2"],ut="\\nicefrac{\\text{d}C}{\\text{d}t}")
    learned = print_pde(xi_learned_means[i][:,np.newaxis],["C","C^2","C^3","C^4"],ut="\\nicefrac{\\text{d}C}{\\text{d}t}")

    if i == 0:
        xi_MSE = "-"
    else:
        xi_MSE = str(round(np.linalg.norm(xi_learned_means[i]-xi_learned_means[i-1]),3))
    
    MF_str = " & $" + MF + "$ ("+ str(round(MF_MSE[i],4)) +")"
    learned_str = " & $" + learned + "$ ("+ str(round(learned_MSE[i],4)) +") & "+xi_MSE+" \\\\"
    
    print("    " + str(real) + MF_str + learned_str )
    print("    \\hline")
print('\\end{tabular}')
```
