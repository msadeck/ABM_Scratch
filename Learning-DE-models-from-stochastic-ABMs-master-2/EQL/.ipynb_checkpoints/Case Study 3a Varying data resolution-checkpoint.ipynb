```python
from PDEFind_class_online import PDE_Findclass
import time, glob, os, pdb
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import set_matplotlib_formats
set_matplotlib_formats('png', 'pdf')
from model_selection_IP3 import ODE_sim,learned_RHS_ODE
from PDE_FIND3 import print_pde

#increase font size throughout
font = {'size'   : 20}
plt.rc('font', **font)
```

```python
# growth, death terms
rp = 0.05
rd = rp/4.0

dataset = ["logistic_ABM_sim_rp_"+str(rp)+"_rd_"+str(rd)+"_real50_samples_13"]
```

```python
#EQL Pipeline values
#Where data is stored
data_dir = "../data/"
#Additional name in filenames
comp_str = ""
model_name = ""

#percentage of training, testing points
trainPerc = 0.5
valPerc = 0.5

#degree of library
deg = 4

#Learning algorithm
algo_name = 'Greedy'

#pruning percentage
prune_level = 0

num_eqns = 3

#number of equations to learn
reals = 10
```

## Perform EQL Pipeline on one dataset

```python
pf = PDE_Findclass(dataset,
                   comp_str,
                   model_name,
                   data_dir=data_dir,
                   reals=reals,
                   trainPerc = trainPerc,
                   valPerc = valPerc,
                   deg=deg,
                   prune_level = prune_level,
                   num_eqns=num_eqns,
                   algo_name = algo_name)

#Solve for xi for $real train-val splits
pf.train_val_ODEFind()

print("")

#determine which equation form is most-commonly recovered
pf.list_common_eqns()
```

## Plot Learned equation against ABM data

```python
#Learned model
y_inferred = ODE_sim(pf.inferred_xi[0],learned_RHS_ODE,pf.t,np.array([0.05]),description=pf.description,deg=deg)
#Mean-field Model
y_MF = ODE_sim(np.array([rp-rd,-rp]),learned_RHS_ODE,pf.t,np.array([0.05]),description=['C^1','C^2'],deg=2)

#plot
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(pf.t*(rp-rd),y_MF.T,'k',label = "MF Model")
ax.plot(pf.t*(rp-rd),y_inferred.T,'r--',label="Learned Model")
ax.plot(pf.t*(rp-rd),pf.y,'b.',label = "ABM Data",markersize=8)

#axes
plt.legend(loc=4)
ax.set_title("$P_p$ = " +str(rp)+", $P_d$ = " +str(rd))
ax.set_xlabel("Time ($T$, Nondimensionalized)")
ax.set_ylabel("Density")
```

## Now do the same for several rp values

```python
# growth, death terms
fontsize= 15
#resolution of data (number of samples)
samples = [100,50,25,13]

rp = 0.05
rd = rp/4.0

fig = plt.figure(figsize=(9,6.5))
count = 1

title_strs = ['$n=100$ time samples',
              '$n=50$ time samples',
              '$n=25$ time samples',
              '$n=13$ time samples']

## Initialize summary lists list of learned xi vectors
xi_learned = []
#list of mean squared error (MSE) between Mean-field model and ABM data
MF_MSE = []
#list of mean squared error (MSE) between learned model and ABM data
learned_MSE = []


for i,sample in enumerate(samples):

    dataset = ["logistic_ABM_sim_rp_"+str(rp)+"_rd_"+str(rd)+"_real50_samples_"+str(sample)]
    
    #run EQL Pipeline
    pf = PDE_Findclass(dataset,
                       comp_str,
                       model_name,
                       data_dir=data_dir,
                       reals=reals,
                       trainPerc = trainPerc,
                       valPerc = valPerc,
                       deg=deg,
                       prune_level = prune_level,
                       num_eqns=num_eqns,
                       algo_name = algo_name)

    pf.train_val_ODEFind()
    pf.list_common_eqns()

    #save fine time grid for simulating learned equation
    if i == 0: t = pf.t
    
    #solve learned model
    y_inferred = ODE_sim(pf.inferred_xi[0],learned_RHS_ODE,t,np.array([0.05]),description=pf.description,deg=deg)
    
    #plot
    ax = fig.add_subplot(2,2,count)
    
    ax.plot(pf.t*(rp-rd),pf.y,'b.-',label = "ABM Data",markersize=7,linewidth=2)
    ax.plot(t*(rp-rd),y_inferred.T,'r--',label="Learned Model")
    
    #axes
    ax.set_title(title_strs[i])
    if count > 2:
        ax.set_xlabel("Time ($T$)",fontsize=fontsize)
    else:
        ax.set_xticks([])
    if count % 2 == 1:
        ax.set_ylabel("Density",fontsize=fontsize)
    ax.set_ylim(0,0.8)
    
    
    # Save MSE, learned xi value    
    learned_MSE.append(np.linalg.norm(np.squeeze(y_inferred[0,::2**i]) - pf.y)/len(pf.y))
    xi_learned.append(pf.inferred_xi[0])    
        
    count+=1
    
    
    
#plt.savefig("case_study_3_BMD_comparison_fewer_time.pdf",format="pdf")
```

## Table of inferred Equations

```python
##Creates table that can be copied and pasted into a LaTeX table environment.

print('\\begin{tabular}{|c|l|}')
print('    \\hline')
print('    \% of data & \\multicolumn{1}{c|}{Learned Model (MSE)} \\\\ ')
print('    \\hline')

divs = ['100','50','25','13']

for i in range(len(divs)):
    rd = rp/2.0

    learned = print_pde(xi_learned[i][:,np.newaxis],["C","C^2","C^3","C^4"],ut="\\nicefrac{\\text{d}C}{\\text{d}t}")
    
    print("    $" + divs[i] + "$ & $" + learned +"$ ("+str(round(learned_MSE[i],4))+") \\\\")
    print("    \\hline")
print('    \\end{tabular}')
```
