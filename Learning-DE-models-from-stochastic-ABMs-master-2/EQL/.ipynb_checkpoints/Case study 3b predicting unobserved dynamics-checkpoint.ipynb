```python
from PDEFind_class_online import PDE_Findclass
import time, glob, os, pdb
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import set_matplotlib_formats
set_matplotlib_formats('png', 'pdf')
from model_selection_IP3 import ODE_sim,learned_RHS_ODE
from PDE_FIND3 import print_pde

#increase font size throughout
font = {'size'   : 15}
plt.rc('font', **font)
```

```python
# growth, death terms
rp = 0.01
rd = rp/4.0

dataset = ["logistic_ABM_sim_rp_"+str(rp)+"_rd_"+str(rd)+"_real50"]
```

```python
#PDE-FIND values
#Where data is stored
data_dir = "../data/"
#Additional name in filenames
comp_str = ""
model_name = ""

#percentage of training, testing points
trainPerc = 0.5
valPerc = 0.5

#degree of library
deg = 4

#Learning algorithm
algo_name = 'Greedy'

#pruning percentage
prune_level = 0

num_eqns = 3
#save_learned_eqns = True
#save_xi = True
#save_learned_xi = True

#number of equations to learn
reals = 10
```

## Perform EQL Pipeline on one dataset

```python
#factor of the data to train on (div_factor = 3 means we will train on the first
#third of data and try and predict the last two-thirds)

div_factor = 3
model_name = "predicting_"+str(div_factor)

pf = PDE_Findclass(dataset,
                   comp_str,
                   model_name,
                   data_dir=data_dir,
                   reals=reals,
                   trainPerc = trainPerc,
                   valPerc = valPerc,
                   deg=deg,
                   prune_level = prune_level,
                   num_eqns=num_eqns,
                   algo_name = algo_name)


#Solve for xi for $real train-val splits
pf.train_val_ODEFind_OOS(div_factor)

print("")

#determine which equation form is most-commonly recovered
pf.list_common_eqns()
```

## Plot learned equation against seen and unseen ABM data

```python
y_inferred = ODE_sim(pf.inferred_xi[0],learned_RHS_ODE,pf.t,np.array([0.05]),description=pf.description,deg=deg)
y_MF = ODE_sim(np.array([rp-rd,-rp]),learned_RHS_ODE,pf.t,np.array([0.05]),description=['C^1','C^2'],deg=2)

y_inferred[y_inferred==0] = np.nan

fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(pf.t*(rp-rd),y_MF.T,'k',label = "MF Model")
ax.plot(pf.t*(rp-rd),y_inferred.T,'r--',label="Learned Model")

N = len(pf.t)//div_factor

ax.plot(pf.t[:N:4]*(rp-rd),pf.y[:N:4],'b.',label = "ABM training Data",markersize=8)
ax.plot(pf.t[N::4]*(rp-rd),pf.y[N::4],'g*',label = "ABM testing Data",markersize=8)

plt.legend(loc=4)

ax.set_title("$P_p$ = " +str(rp)+", $P_d$ = " +str(rd))
ax.set_xlabel("Time ($T$, Nondimensionalized)")
ax.set_ylabel("Density")

```

## Now do the same for several rp values

```python
# how much of the data to train on (one-tenth, one-fifth, one-fourth, one-half)
div_factors = [10,5,4,2]#

rp = 0.01
rd = rp/4.0

dataset = ["logistic_ABM_sim_rp_"+str(rp)+"_rd_"+str(rd)+"_real50"]

fig = plt.figure(figsize=(9,6.5))
count = 1

title_strs = ['Trained on 10% of data','Trained on 20% of data','Trained on 25% of data','Trained on 50% of data']

## Initialize summary lists list of learned xi vectors
xi_learned = []
#list of mean squared error (MSE) between Mean-field model and ABM data
MF_MSE = []
#list of mean squared error (MSE) between learned model and ABM data
learned_MSE = []


for i,div_factor in enumerate(div_factors):

    model_name = "predicting_"+str(div_factor)

    #run EQL Pipeline    
    pf = PDE_Findclass(dataset,
                   comp_str,
                   model_name,
                   data_dir=data_dir,
                   reals=reals,
                   trainPerc = trainPerc,
                   valPerc = valPerc,
                   deg=deg,
                   prune_level = prune_level,
                   num_eqns=num_eqns,
                   algo_name = algo_name)



    pf.train_val_ODEFind_OOS(div_factor)
    pf.list_common_eqns()

    
    #simulated learn model
    y_inferred = ODE_sim(pf.inferred_xi[0],learned_RHS_ODE,pf.t,np.array([0.05]),description=pf.description,deg=deg)
    #helps in plotting learned equation for first learning case
    y_inferred[y_inferred==0] = np.nan

    #unobserved data
    y_test = pf.y
    y_test = y_test[len(y_test)//div_factor:]
    
    #plot
    ax = fig.add_subplot(2,2,count)
    N = len(pf.t)//div_factor
    
    ax.plot(pf.t[:N:4]*(rp-rd),pf.y[:N:4],'b.-',label = "ABM training Data",markersize=8)
    ax.plot(pf.t[N::4]*(rp-rd),pf.y[N::4],'g*-',label = "ABM testing Data",markersize=6)
    ax.set_title(title_strs[i])#, $P_d$ = " +str(rd)
    ax.plot(pf.t*(rp-rd),y_inferred.T,'r--',label="Learned Model")
    
    #axes
    if count > 2:
        ax.set_xlabel("Time ($T$)")
    else:
        ax.set_xticks([])
    if count % 2 == 1:
        ax.set_ylabel("Density")
    ax.set_ylim(0,1.0)
    
    
    # Save MSE, learned xi value    
    learned_MSE.append(np.linalg.norm(y_test - y_inferred[0,len(pf.y)//div_factor:])/len(y_test))
    xi_learned.append(pf.inferred_xi[0])
    
    
    count+=1
    

    
#plt.savefig("case_study_3_BMD_comparison.pdf",format="pdf")
```
